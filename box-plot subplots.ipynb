```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import math
from matplotlib.ticker import FormatStrFormatter, MultipleLocator 
```

```python
data = pd.read_csv('C:/Users/KarimovDO/Desktop/базы/gepatit.csv')
group = ['Casp7', 'Chek', 'Gclc','Gstm1', 'Gstp1', 'Gstt1', 'Hmox', 'Nfe2l2', 'Nqo1', 'Ripk', 'Sod1']
```

```python
for i in range(len(data['group'])):
    if data.loc[i,'group'] == 0:
        data.loc[i,'group'] = 'K-'
    if data.loc[i,'group'] == 1:
        data.loc[i,'group'] = 'K+'
    if data.loc[i,'group'] == 2:
        data.loc[i,'group'] = 'Geptor'
    if data.loc[i,'group'] == 3:
        data.loc[i,'group'] = 'Mexidol'
    if data.loc[i,'group'] == 4:
        data.loc[i,'group'] = 'OMU'
        
data['group'].unique()
```

```python
data = data[data['toxicant'] == 1]
```

```python
data.info()
```

```python
sns.set_theme(style="whitegrid")
data_vr = data[data['toxicant'] == 1]
k = 0
f, ax = plt.subplots(3, 4, figsize=(16, 10), sharex=True)
for i in range(3):
    for j in range(4):
        sns.boxplot(data=data_vr, x="group", y=group[k], hue="time", notch=True,  palette=["b",".85"], ax = ax[i,j])
        ax[i,j].set_xlabel("")
        ax[i,j].yaxis.set_major_formatter(FormatStrFormatter('%d'))
        ax[i,j].yaxis.set_major_locator(MultipleLocator(base=1))
        ax[i,j].get_legend().remove()
        k += 1
        if k>10:
            k=10
ax[2,3].set_ylim(19, 20)
ax[2,3].set_ylabel("")
ax[2,3].legend(fontsize = 36)
plt.savefig('C:/Users/KarimovDO/Desktop/базы/tcc.png', dpi = 1200)
```
